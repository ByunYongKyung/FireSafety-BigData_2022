### 트레인값만 스케일러에 학습

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(feature, target, test_size=0.20, shuffle=True,stratify = target ,random_state=34)
```

### 스케일러 불러오기(RobustScaler)

```python
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler

scaler = RobustScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_valid_scaled = scaler.transform(x_valid)
```

###keras 에서 신경망 층을 쌓기 위한 라이브러리

```python
from keras.layers import Input, Dense, BatchNormalization, Dropout
from keras.models import Model
```

##모델 만들기

```python
input = Input(shape = (x_train_scaled.shape[1],))
x = Dense(512,activation = 'selu')(input)
x = Dense(256,activation = 'selu')(x)
x = Dropout(0.2)(x)
x = Dense(256,activation = 'selu')(x)
x = Dropout(0.2)(x)
x = Dense(128,activation = 'selu')(x)


x = Dropout(0.2)(x)
x = Dense(128,activation = 'selu')(x)
x = Dense(64,activation = 'selu')(x)
x = Dense(64,activation = 'selu')(x)
x = Dense(32,activation = 'selu')(x)
x = Dense(16,activation = 'selu')(x)
output = Dense(target.nunique(),activation = 'softmax')(x)

model = Model(inputs=input, outputs=output)
```

```python
model.summary()
```

###모델 CALLBACKS

#### Learning Schedular

```python
import tensorflow as tf
def step_decay(epoch,lr):
  if epoch < 900:
    return lr
  else:
    return lr * tf.math.exp(-0.05)

schedule = tf.keras.callbacks.LearningRateScheduler(
    schedule =step_decay , verbose=0
)
```

####모델 체크포인트(최고 성능 모델 저장)

```python
from keras.callbacks import ModelCheckpoint

filename = './chkpnt/3번째모델/{epoch:04d}.ckpt'
checkpoint = ModelCheckpoint(filename,             # file명을 지정합니다
                             monitor='val_accuracy',   # val_loss 값이 개선되었을때 호출됩니다
                             verbose=1,            # 로그를 출력합니다
                             save_best_only=True,  # 가장 best 값만 저장합니다
                             mode='auto'           # auto는 알아서 best를 찾습니다. min/max
                            )
```

####모델 컴파일

##### optimizer = Nadam, sgd

```python
model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
```
