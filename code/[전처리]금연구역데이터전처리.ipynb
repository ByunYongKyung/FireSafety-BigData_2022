```python
nodamZone_gu_df = pd.DataFrame(nodamZone_df['시군구명'].value_counts()).reset_index()
nodamZone_gu_df.rename(columns={'index':'행정구역', '시군구명':'nodamZone'}, inplace=True)
```

```python
sample_df = pd.read_csv('14~16년도 각자 데이터 채우기.csv', encoding='cp949')
sample_df.tail(4)
```

```python
fire_total_nodam_Ver = pd.merge(sample_df, nodamZone_gu_df, how='outer', on=['행정구역'])
fire_total_nodam_Ver.head()
```
