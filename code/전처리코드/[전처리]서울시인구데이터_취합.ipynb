```python
popul_simple_df = popul_df.iloc[66:,:]
popul_simple_df.tail()
```

```python
totalPopul_df = popul_simple_df[(popul_simple_df['항목'] == '계') & (popul_simple_df['연령별'] == '계')]
totalPopul_df.drop(columns=['연령별', '항목', '단위'], inplace=True)
```

```python
oldAge_df = popul_simple_df[(popul_simple_df['항목'] == '계') & ((popul_simple_df['연령별'] == '70~74세') |
                           (popul_simple_df['연령별'] == '75~79세') | (popul_simple_df['연령별'] == '80~84세') |
                           (popul_simple_df['연령별'] == '85~89세') | (popul_simple_df['연령별'] == '90~94세') |
                           (popul_simple_df['연령별'] == '95~99세') | (popul_simple_df['연령별'] == '100세 이상'))]
oldAge_df.drop(columns=['연령별', '항목', '단위'], inplace=True)
```

```python
oldAge_df_tmp = pd.DataFrame(index = oldAge_df['동별'].unique(), columns=oldAge_df.columns[1:])
oldAge_df_tmp.head(3)
```

```python
# 통합데이터에 merge 하기위한 함수
def fd_precessing(df, col_name):
    
    df.rename(columns = {'동별':'행정구역'}, inplace=True)
    df = df.set_index('행정구역')

    concatVar = []

    for i in range(len(df.columns)):
        globals()[f'sp_df_{i}']=df.iloc[:,i:i+1]
        concatVar.append(globals()[f'sp_df_{i}'])
        concatVar[i].reset_index(inplace=True)
        concatVar[i]['집계년월'] = concatVar[i].columns[1]
        concatVar[i].rename(columns={concatVar[i].columns[1] : col_name}, inplace=True)


    return pd.concat(concatVar, axis=0)
```

```python
totalPopul_df = fd_precessing(totalPopul_df, 'totalPopul')
totalPopul_df
```

```python
oldAgePopul_df = fd_precessing(oldAge_df, 'oldAgePopul')
oldAgePopul_df.head()
```

```python
population_df = pd.merge(totalPopul_df[totalPopul_df['행정구역'] != '금천구'], oldAgePopul_df[oldAgePopul_df['행정구역'] != '금천구'], how='outer', on=['집계년월', '행정구역'])
# population_df.to_csv('fire_total_popul_test.csv')
population_df
```

```python
fire_total_popul_Ver = pd.merge(sample_df, population_df, how='outer', on=['집계년월', '행정구역'])
fire_total_popul_Ver.head()
```

```python
fire_total_popul_Ver['totalPopul'] = fire_total_popul_Ver['totalPopul'].fillna(method='ffill')
fire_total_popul_Ver['oldAgePopul'] = fire_total_popul_Ver['oldAgePopul'].fillna(method='ffill')
```
